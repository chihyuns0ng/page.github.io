```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
df = sns.load_dataset("iris")
df.shape
```

```python
df.info()
```

```python
df.head()
```

```python
df["species"]
```

```python
_ = sns.boxplot(data = df)
```

```python
_ = sns.countplot(data=df, x="species")
```

```python
df.groupby("species")["petal_length"].mean()
```

```python
_ = sns.barplot(data=df, x="petal_width", y="petal_length", hue="species", ci=None)
```

```python
_ = sns.catplot(data=df, x="species", y="sepal_width", kind="violin")
```

```python
df.nunique()
```

```python
plt.figure(figsize=(10,6))
cat_kind=["box", "violin", "strip", "boxen"]
iris = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
for i in cat_kind:
    for j in iris:
        _ = sns.catplot(data=df, x="species", y=j, kind=i)
```
